Normalized:

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Provide the file path
file_path = r"C:\---Address file -----\Book1.xlsx"

# Read the Excel file using pandas
df = pd.read_excel(file_path)

# Extract the first column (assuming it's the index)
index_col = df.columns[0]

# Normalize data except the first column
scaler = MinMaxScaler(feature_range=(0, 1))
df.iloc[:, 1:] = scaler.fit_transform(df.iloc[:, 1:])

# Scale data to have a minimum of 0 and a maximum of 30
df.iloc[:, 1:] = df.iloc[:, 1:] * 31

# Plotting
plt.figure(figsize=(10, 6))

# Plot each column except the first one
for column in df.columns[1:]:
    plt.plot(df[index_col], df[column], label=column)

plt.xlabel(index_col)
plt.ylabel('Scaled Values')
plt.title('Normalized and Scaled Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Add small random noise to each curve
noise = np.random.normal(0, 0.22, size=df.shape[0])
df['speed (knots)'] += noise
df['speed (knots) 2'] += noise
df['speed (knots) 3'] += noise
df['speed (knots) 4'] += noise

# Set specific values to zero for the given time ranges
df.loc[(df['Time [H]'] >= 0) & (df['Time [H]'] <= 0.5), ['speed (knots)', 'speed (knots) 2', 'speed (knots) 3', 'speed (knots) 4']] = 0
df.loc[(df['Time [H]'] >= 10) & (df['Time [H]'] <= 12), ['speed (knots)', 'speed (knots) 2', 'speed (knots) 3', 'speed (knots) 4']] = 0
df.loc[(df['Time [H]'] >= 22.5) & (df['Time [H]'] <= 25), ['speed (knots)', 'speed (knots) 2', 'speed (knots) 3', 'speed (knots) 4']] = 0

# Plot the data
plt.figure(figsize=(10, 6))

# Assuming 'Time [H]' is the x-axis, and 'speed (knots)' and its variations are y-axes
plt.plot(df['Time [H]'], df['speed (knots)'], label='Speed 1 - Probability = 0.4')
plt.plot(df['Time [H]'], df['speed (knots) 2'], label='Speed 2 - Probability = 0.3')
plt.plot(df['Time [H]'], df['speed (knots) 3'], label='Speed 3 - Probability = 0.18')
plt.plot(df['Time [H]'], df['speed (knots) 4'], label='Speed 4 - Probability = 0.12')

# Set labels and title
plt.xlabel('Time [H]', fontsize=20)
plt.ylabel('Speed (knots)', fontsize=20)
#plt.title('Speed vs Time', fontsize=20)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)



# Display legend
plt.legend()
plt.legend(fontsize=14)


# Save as PDF
plt.savefig('plot.pdf')

```

Convert Speed to Power By Interpolation:

```python
import pandas as pd
from scipy.interpolate import interp1d

# Define the file path for the Excel file
file_path3 = r"C:\---Address file -----\S2P.xlsx"

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path3)

# Extract speed and power data from DataFrame
speed = df.iloc[:, 0]  # Assuming speed is in the first column
power = df.iloc[:, 1]  # Assuming power is in the second column

# Create interpolation function
interp_func = interp1d(speed, power, kind='linear')

# Function to predict power for any given speed
def predict_power(speed_value):
    return interp_func(speed_value)

# Test the function
speed_value = 26  # Example speed value
predicted_power = predict_power(speed_value)
print(f"Predicted power at {speed_value} knots: {predicted_power} kW")
```
